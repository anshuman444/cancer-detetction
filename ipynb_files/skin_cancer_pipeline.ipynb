# 🧬 Skin Cancer detection: Transfer Learning Pipeline

This notebook presents the final pipeline for skin lesion classification, utilizing **Transfer Learning** to distinguish between **Malignant** and **Benign** lesions.

## 1. Environment Setup

```python
import tensorflow as tf
from tensorflow.keras.models import load_model
from PIL import Image, ImageOps
import numpy as np
import matplotlib.pyplot as plt
import json
import os

print("Libraries loaded.")
```

## 2. Preprocessing Core

Standardizing dermal images ensures variant lighting conditions do not impact the diagnostic result.

```python
def clean_lesion_image(image_path):
    img = Image.open(image_path).convert('RGB')
    img_resized = ImageOps.fit(img, (224, 224), Image.Resampling.LANCZOS)
    img_array = np.asarray(img_resized) / 255.0
    return img_resized, np.expand_dims(img_array, axis=0)

print("Skin preprocessing pipeline active.")
```

## 4. Fine-Tuning Analytics (10 Epochs)

Using a pre-trained base, the model achieved domain-specific expertise in just **10 epochs**.

```python
history_file = '../json_files/skin cancer/training_history.json'
with open(history_file, 'r') as f:
    history = json.load(f)

epochs = range(1, len(history['accuracy']) + 1)

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(epochs, history['accuracy'], 'orange', label='Acc')
plt.plot(epochs, history['val_accuracy'], 'black', label='Val Acc')
plt.title('Skin Model Learning Progress')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(epochs, history['loss'], 'orange', label='Loss')
plt.plot(epochs, history['val_loss'], 'black', label='Val Loss')
plt.title('Skin Model Error Rate')
plt.legend()
plt.show()
```

## 5. Risk Assessment (Prediction)

Predictive analysis for high-risk melanoma vs benign nevi.

```python
model_path = '../models/skin_cancer_model.h5'
skin_model = load_model(model_path)

def assesses_skin_lesion(image_path):
    _, model_input = clean_lesion_image(image_path)
    prediction = skin_model.predict(model_input)
    
    # Binary classification: 0=Benign, 1=Malignant
    idx = np.argmax(prediction, axis=1)[0]
    score = prediction[0][idx] * 100
    status = "🚨 MALIGNANT (Risk Detected)" if idx == 1 else "✅ BENIGN"
    
    print(f"Skin Diagnostic Result: {status}")
    print(f"AI Confidence Score: {score:.2f}%")

print("Diagnostic engine ready.")
skin_model.summary()
```
